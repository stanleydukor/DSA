### Singly LinkedList

Linked lists excel in terms of space complexity

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.length = 0
    
    def append(self, value): # O(n)
        if not self.head:
            self.head = Node(value)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(value)
        self.length += 1

    def prepend(self, value): # O(1)
        current = self.head
        self.head = Node(value)
        self.head.next = current
        self.length += 1

    def insert (self, value, idx):
        current = self.head
        if idx == 0:
            self.head = Node(value)
            self.head.next = current
        else:
            current_idx = 0
            while current:
                if current_idx + 1 == idx:
                    previous = current.next
                    current.next = Node(value)
                    current.next.next = previous
                    break
                else:
                    current = current.next
                    current_idx += 1
        self.length += 1

    def remove(self, idx):
        current = self.head
        if idx == 0:
            self.head = self.head.next
        else:
            current_idx = 0
            while current:
                if current_idx + 1 == idx:
                    current.next = current.next.next
                    break
                else:
                    current = current.next
                    current_idx += 1
        self.length -= 1
    
    def display(self): # O(n)
        arr = []
        current = self.head
        while current:
            arr.append(current.data)
            current = current.next
        return arr
    
    def search(self, value): # O(n)
        current = self.head
        while current:
            if current.data == value:
                return True
            current = current.next
        return False
    
    def __len__(self):
        return self.length
    

    # insert # O(n)
    # remove # O(n)
    # doubly linked list
    # reverse
```

```python
linkedList = LinkedList()
linkedList.append(1)
linkedList.append(2)
linkedList.append(3)
linkedList.append(4)
linkedList.append(5)
print(linkedList.display())

linkedList.prepend(6)
print(linkedList.display())

linkedList.insert(100, 4)
print(linkedList.display())

linkedList.remove(0)
linkedList.remove(3)
print(linkedList.display())

print(linkedList.search(6))
print(len(linkedList))
```

### Doubly LinkedList

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = Node(0)
        self.tail = Node(0)
        self.head.next = self.tail
        self.tail.prev = self.head

    def append(self, value):
        newNode = Node(value)
        newNode.next = self.tail
        newNode.prev = self.tail.prev
        newNode.prev.next = newNode
        self.tail.prev = newNode

    def prepend(self, value):
        newNode = Node(value)
        newNode.prev = self.head
        newNode.next = self.head.next
        newNode.next.prev = newNode
        self.head.next = newNode

    def insert(self, value, idx):
        pass

    def remove(self, idx):
        pass

    def search(self, value):
        pass
    
    def displayForward(self):
        arr = []
        current = self.head
        while current:
            arr.append(current.value)
            current = current.next
        return arr
    
    def displayForward(self):
        array = []
        current = self.head.next
        while current.next:
            array.append(current.value)
            current = current.next
        return array
    
    def displayBackward(self):
        array = []
        current = self.tail.prev
        while current.prev:
            array.append(current.value)
            current = current.prev
        return array        
```

```python
linkedList = DoublyLinkedList()
linkedList.append(1)
linkedList.append(2)
linkedList.append(3)
linkedList.append(4)
linkedList.append(5)
print(linkedList.displayForward())
print(linkedList.displayBackward())
```

### Queue

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if len(self.items) != 0:
            self.items.pop(0)

    def display(self):
        return self.items
    
    def __len__(self):
        return len(self.items)
```

```python
queue = Queue()
queue.enqueue('a')
queue.enqueue('b')
queue.enqueue('c')
print(queue.display())

queue.dequeue()
print(queue.display())
```

### Graph Traversal

```python
# BFS
from collections import deque

def bfs(graph, start):
    queue = deque([start])
    visited = set([start])
    while queue:
        vertex = queue.popleft()
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return visited

def dfs_recursive(graph, vertex, visited=None):
    if visited == None:
        visited = set()
    visited.add(vertex)
    for neighbor in graph[vertex]:
        if neighbor not in visited:
            visited = dfs_recursive(graph, neighbor, visited)
    return visited

def dfs_iterative(graph, start):
    stack = [start]
    visited = set([start])
    while stack:
        vertex = stack.pop()
        for neighbor in reversed(graph[vertex]):
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
    return visited
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C']
}

print(bfs(graph, 'A'))
print(dfs_recursive(graph, 'A'))
print(dfs_iterative(graph, 'A'))
```

### Merge Sort

```python
def mergeSort(arr):
    if len(arr) > 1:
 
         # Finding the mid of the array
        mid = len(arr)//2
 
        # Dividing the array elements
        L = arr[:mid]
 
        # Into 2 halves
        R = arr[mid:]
 
        # Sorting the first half
        mergeSort(L)
 
        # Sorting the second half
        mergeSort(R)
 
        i = j = k = 0
 
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
 
        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
 
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
 
 
# Code to print the list
def printList(arr):
    for i in range(len(arr)):
        print(arr[i], end=" ")
    print()
 
 
arr = [12, 11, 13, 5, 6, 7]
print("Given array is")
printList(arr)
mergeSort(arr)
print("\nSorted array is ")
printList(arr)
```

### Binary Tree

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def add(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self.add_recursive(self.root, value)

    def add_recursive(self, node, value):
        if value < node.value:
            if node.left:
                self.add_recursive(node.left, value)
            else:
                node.left = Node(value)
        else:
            if node.right:
                self.add_recursive(node.right, value)
            else:
                node.right = Node(value)

    def traverse(self):
        return self.traverse_recursive(self.root, [])

    def traverse_recursive(self, node, result):
        if node:
            self.traverse_recursive(node.left, result)
            result.append(node.value)
            self.traverse_recursive(node.right, result)
        return result
```

```python
bt = BinaryTree()
bt.add(5)
bt.add(3)
bt.add(7)
bt.add(2)
bt.add(4)
bt.add(6)
bt.add(8)
print("Inorder traversal of the binary tree:")
print(bt.traverse())
```
