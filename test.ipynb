```python
def recursive_merge(intervals):
    if len(intervals) > 1:
        mid = len(intervals) // 2
        L = intervals[:mid]
        R = intervals[mid:]

        mergedL = recursive_merge(L)
        mergedR = recursive_merge(R)

        if mergedL[-1][-1] >= mergedR[0][0]:
            new_intervals = mergedL[:-1] + [[min(mergedL[-1][0], mergedR[0][0]), max(mergedL[-1][1], mergedR[0][1])]] + mergedR[1:]
        else:
            new_intervals = mergedL + mergedR

        return new_intervals

    return intervals

def merge_sort(intervals):
    if len(intervals) > 1:
        mid = len(intervals) // 2
        L = intervals[:mid]
        R = intervals[mid:]

        merge_sort(L)
        merge_sort(R)
        
        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i][0] < R[j][0]:
                intervals[k] = L[i]
                i += 1
            else:
                intervals[k] = R[j]
                j += 1
            k += 1
            
        while i < len(L):
            intervals[k] = L[i]
            i += 1
            k += 1
            
        while j < len(R):
            intervals[k] = R[j]
            j += 1
            k += 1

def merge(intervals):
    merge_sort(intervals)
    print(intervals)
    results = recursive_merge(intervals)
    return results
```
